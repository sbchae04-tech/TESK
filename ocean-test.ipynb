```python
from dwave.system import LeapHybridSampler

# QUBO:
# E(x0, x1) = -x0 - x1 + 2 x0 x1
#
# 가능한 값:
# x0 x1 = 00 -> E = 0
# x0 x1 = 10 -> E = -1
# x0 x1 = 01 -> E = -1
# x0 x1 = 11 -> E = 0
#
# 따라서 최적해는 (1,0) 또는 (0,1)

Q = {
    ("x0", "x0"): -1,
    ("x1", "x1"): -1,
    ("x0", "x1"): 2,
}

sampler = LeapHybridSampler()

sampleset = sampler.sample_qubo(Q, label="Simple QUBO Example")

print(sampleset)
print(sampleset.first)
```
