```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0.0, 1.4, 1000)
b = 0.3
```

```python
x0 = y0 = 0
```

```python
N_total = 100000
N_trash = 1000

x = np.zeros(N_total)
y = np.zeros(N_total)

x[0] = x0
y[0] = y0
```

```python
a_plot = []
x_plot = []

for a in a:
    x = x0
    y = y0

    for n in range(N_total):
        x_next = 1 - a * x**2 + y
        y_next = b * x

        x = x_next
        y = y_next

        # transient 이후의 x 값만 저장
        if n >= N_trash:
            a_plot.append(a)
            x_plot.append(x)
```

```python
plt.figure(figsize=(7, 6), dpi=150)
plt.scatter(a_plot, x_plot, s=0.1, label=f"b={b} bifurcation diagram")
plt.xlabel(r"$a$")
plt.ylabel(r"$x_n$")
plt.title(r"Hénon Map")
plt.grid(False)
plt.legend(loc="upper right")
plt.axis("equal")
plt.savefig("graph_Bif.png", dpi=300, bbox_inches="tight")
plt.show()
```
