```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1.4
b = 0.3
```

```python
x0 = y0 = 0
```

```python
N_total = 100000
N_trash = 1000

x = np.zeros(N_total)
y = np.zeros(N_total)

x[0] = x0
y[0] = y0
```

```python
for n in range(N_total - 1):
    x[n+1] = 1 - a * x[n]**2 + y[n]
    y[n+1] = b * x[n]

x_plot = x[N_trash:]
y_plot = y[N_trash:]
```

```python
plt.figure(figsize=(7, 6), dpi=150)
plt.scatter(x_plot, y_plot, s=0.1, label=f"a={a}, b={b}")
plt.xlabel(r"$x_n$")
plt.ylabel(r"$y_n$")
plt.title(r"Hénon Map")
plt.grid(False)
plt.legend(loc="upper right")
plt.axis("equal")
plt.savefig("graph_HenonMap.png", dpi=300, bbox_inches="tight")
plt.show()
```
