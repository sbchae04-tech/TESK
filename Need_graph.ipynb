```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Function(x, a):
    return (x - a)**2
```

```python
x = np.linspace(-5, 5, 200)
a_1 = 0
a_2 = 2

F_x_1 = Function(x, a_1)
F_x_2 = Function(x, a_2)
```

```python
plt.plot(x, F_x_1, label = r"$F = x^2$")
plt.plot(x, F_x_2, label = r"$F =x^2-4x+4  $")
plt.axvline(x = a_1)

plt.xlim(-5, 5)
plt.xlabel("x")
plt.ylabel("F(x)")

plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

s = np.linspace(-5, 5, 200)

F1 = s**2
F2 = (s - 2)**2
F3 = s**2 + 0.5 * (s - 2)**2

plt.figure(figsize=(7, 5), dpi=150)

plt.plot(s, F1, label=r"$F_1=s^2$")
plt.plot(s, F2, label=r"$F_2=(s-2)^2$")
plt.plot(s, F3, label=r"$F_3=s^2+\frac{1}{2}(s-2)^2$")

plt.xlabel("s")
plt.ylabel("F(s)")
plt.grid(True)
plt.legend()
plt.show()
```
